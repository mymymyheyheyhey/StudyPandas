## 데이터 병합

```python
dict_s1 = {'A' : 0, 'B' : 1}
s1 = pd.Series(dict_s1)
s1
```

```python
dict_s2 = {'a' : 2, 'b' : 3, 'c' : 4}
s2 = pd.Series(dict_s2)
s2
```

```python
# 세로 병합
pd.concat([s1, s2], axis = 0)
```

### 컬럼 단위로 병합

```python
colnames = ["동물", "나이"]
df_temp1 = pd.DataFrame([ ['Dog', 7]
                        , ['Cat', 9]
                        , ['Tiger', 2]
                        , ['Lion', 3]  ]
                        , index = [0, 1, 2, 3]
                        , columns = colnames)
df_temp1
```

```python
colnames2 = ["사는곳", "뿔의 개수"]
df_temp2 = pd.DataFrame([ ['집', 0]
                        , ['초원', 1]
                        , ['수풀', 2]
                        , ['초원', 3]  ]
                        , index = [0, 1, 2, 3]
                        , columns = colnames2)
df_temp2
```

```python
# 가로 병합
pd.concat([df_temp1, df_temp2], axis = 1)
```

```python
mem = {"고객번호" : [1001, 1002, 1003, 1004, 1005, 1006, 1007], "이름" : ['강', '김', '홍', '이', '박', '최', '소']}
colname = ["고객번호", "이름"]

df_info1 = pd.DataFrame(mem, columns = colname)
df_info1
```

```python
mem2 = {"고객번호" : [1001, 1001, 1005, 1006, 1008, 1001]
       , "금액" : [10000, 20000, 15000, 5000, 100000, 30000]}
colname2 = ["고객번호", "금액"]
df_info2 = pd.DataFrame(mem2, columns = colname2)
df_info2
```

```python
# inner join 과 같은 원리
pd.merge(df_info1, df_info2, how = 'inner')
```

```python
# outer join
pd.merge(df_info1, df_info2, how = 'outer')
```

```python
# left outer join
pd.merge(df_info1, df_info2, how = 'left')
```

```python
# right outer join
pd.merge(df_info1, df_info2, how = 'right')
```
