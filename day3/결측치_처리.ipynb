```python
# 데이터 로드
file_path = './data/bicycle.csv'
df = pd.read_csv(file_path) # df = pd.read_csv(file_path, encoding = 'euc-kr')
df
```

```python
# 컬럼별 결측치가 있는 행의 개수 조회
df.isnull().sum(0)
```
