### 데이터 재구조화

```python
df1 = pd.cut(df['math'], 3)
df1.value_counts()
```

```python
df2 = df['math'].groupby(df1)
```

```python
df2.head(1)
```

```python
df2.describe()
```

```python
df2.agg(['count', 'mean', 'std', 'min', 'max'])
```

```python
pd.qcut(df['math'], 3)
```

```python
# qcut 구간을 나누어 유사한 값끼리 묶음
df4 = pd.qcut(df['math'], 3, labels = False)
df4
```

```python
df5 = pd.qcut(df['math'], 3, labels = np.arange(3, 0, -1))
df5
```

```python
df6 = df[['math']].groupby(df5)
df6.head(1)
```

```python
df6.describe()
```

```python
df7 = df6.agg(['count', 'mean', 'std', 'min', 'max'])
df7
```
