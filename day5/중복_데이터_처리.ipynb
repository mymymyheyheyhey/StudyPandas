### 중복 데이터 제거

```python
# 첫번째 중복 데이터만 남기고 나머지 제거
df.drop_duplicates(["이용거리"], keep = 'first') # last 이용 시 마지막 중복 데이터만 남고 모두 제거됨
```

```python
# 중복 데이터 모두 제거
df.drop_duplicates(["이용거리"], keep = False)
```
